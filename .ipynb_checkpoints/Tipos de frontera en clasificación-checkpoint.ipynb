# Tipos de frontera en clasificación

Para estudiar distintos tipos de fronteras, se trabajará con un dataset sintético, el cual se genera a continuación. Este dataset está conformado por dos conjuntos de datos, uno que es generado de una distribución multivariada gaussiana y el otro es generado a través de senos y cosenos. 

```python
# Extraer arreglo de archivos de texto para mantener datos:
# algo así,
#with open("datos1.txt", "r") as ins:
#    datos1 = []
#    for line in ins:
#        array.append(line)
# hacer una lista y luego pasar a np.ndarray(lista_con_comas), implica colocar comas en datos1 para pasarlo asi..
```

```python
from sklearn.utils import check_random_state
import numpy as np

#Cantidad de datos en cada conjunto
n_samples=500

# Media en distribución normal
mean = (0,-4)

# Matriz de Covarianza
C = np.array([[0.3, 0.1], [0.1, 1.5]])

# Distribución Normal Multivariada
datos1 = np.random.multivariate_normal(mean, C, n_samples)

#Senos y cosenos
outer_circ_x = np.cos(np.linspace(0, np.pi, n_samples))*3
outer_circ_y = np.sin(np.linspace(0, np.pi, n_samples))*3
datos2 = np.vstack((outer_circ_x,outer_circ_y)).T
generator = check_random_state(10)
datos2 += generator.normal(scale=0.3, size=datos2.shape)

```

A continuación se presenta una visualización de los datos originales que se obtienen en el dataset sintético.

```python
import matplotlib.pyplot as plt
def visualize(x,y,title=""):
    fig = plt.figure(figsize=(12,6))
    plt.scatter(x[:,0], x[:,1], s=50, c=y, cmap=plt.cm.winter)
    h = .02 # step size in the mesh
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
    plt.title(title)
    plt.show()
```

```python
X_origen = np.concatenate((datos1, datos2), axis=0)
y1 = np.zeros(datos1.shape[0])
y2 = np.ones(datos2.shape[0])
y_origen = np.concatenate((y1,y2),axis=0)
visualize(X_origen,y_origen,"Datos Originales")
```

```python
print len(X_origen)
print len(y_origen)
print len(y1)
print len(y2)
```

De estos datos es posible ver que la gaussiana multivariada está representada por los puntos azules con forma ovalada, mientras que los datos de senos y cosenos están representados por los puntos verdes que forman una semi-luna.

### a) Shift de conjunto 2 al 1 

A continuación se realiza un *shift* entre datos de los conjuntos, es decir, 20 datos del conjunto de datos de la semi-luna pasan a ser parte del conjunto de datos del ovalo. Lo anterior, es posible visualizarlo en el siguiente diagrama, en donde parte de la derecha de los datos verdes pasaron a ser azules. Este *shift* o ruido que se le agrega a los datos se realiza para trabajar con un dataset no ideal y el problema no sea trivial.

Originalmente cada conjunto de datos tiene 500 elementos, pero luego del *shift*, el conjunto de datos verde tiene 480 elementos y el conjunto de datos azules tiene 520 elementos.

```python
X = np.concatenate((datos1, datos2), axis=0)
n = 20 #ruido/noise
y1 = np.zeros(datos1.shape[0]+n)
y2 = np.ones(datos2.shape[0]-n)
y = np.concatenate((y1,y2),axis=0)

visualize(X,y,"Datos con ruido")
```

```python
print "Azules (Multivariada): " + str(len(y1))
print "Verdes (Senos y Cosenos): " + str(len(y2))
```

Para visualizar las fronteras de los distintos algoritmos clasificadores más adelante, se utilizará el siguiente código:

```python
def visualize_border(model,x,y,title=""):
    fig = plt.figure(figsize=(12,6))
    plt.scatter(x[:,0], x[:,1], s=50, c=y, cmap=plt.cm.winter)
    h = .02 # step size in the mesh
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z, cmap=plt.cm.Paired)
    plt.title(title)
    plt.show()
```

### b) Clasificador LDA 

Se entrena un clasificador Linear Discriminant Analysis (LDA), el cual intenta separar los dos conjuntos de datos mediante una combinación lineal de las características.
LDA por definición, supone una distribución normal multivariada para la probabilidad condicionada del dato a cada clase y una matriz de covarianza $\left(\Sigma\right)$ común para todas las clases. De esta manera la frontera entre las clases es lineal.

Se asume que :

<center> $$\begin{equation}
\Sigma_{k} = \Sigma \ \ \ \ \forall k = 1,2,..,K \ \ \ \ (1)
\end{equation}$$
</center>
Luego la función discriminatoria lineal está dada por : 
<center> $$\begin{equation}
\delta_{k}(x) = x^{T} \Sigma^{-1} \mu_{k} - \frac{1}{2} \mu_{k}^{T} \Sigma^{-1} \mu_{k} + log \pi_{k} \ \ \ \ \ (2) 
\end{equation}$$
</center>

Donde:
$$  $$
$\mu_{k}$: corresponde al valor esperado de la distribución normal multivariada para la clase k.                              
$\pi_{k}$: corresponde a la probabilidad de que la clase sea k.

Esta función descrimina al dato y lo clasifica en la clase  que tenga un mayor valor.                       
Debido a la forma de los datos tomado, ya se podría sospechar que este clasificador no funcionaría bien, dado que es prácticamente imposible separar todos los puntos azules de los verdes con una recta.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
model = LDA()
model.fit(X,y)
visualize_border(model,X,y,"LDA con ruido")
```

Se puede observar que mediante LDA, los conjuntos de datos son separado a través de una recta como se mencionó anteriormente. Ejecutando LDA se pudo que ver efectivamente las clases no fueron bien separadas, quedando puntos azul en la zona verde.

```python
from sklearn.metrics import accuracy_score
#Para LDA
y_true = y
y_pred = model.predict(X)
model.predict(X)
print("Miss Classification Loss: %f"%(1-accuracy_score(y_true, y_pred)))
```

Con este error es posible ver que el modelo no se pudo ajustar bien a los datos de entrenamiento que se le dió, teniendo un 2.1% de datos mal clasificados, lo que corresponde a 21 datos, de los cuales ya se sabe que 20 son producto del shift. Este análisis tan detallado sobre la cantidad de datos es posible realizarlo debido a que se está trabajando con un dataset sintético.

```python
model = LDA()
model.fit(X_origen,y_origen)
visualize_border(model,X_origen,y_origen,"LDA sin ruido")

y_true = y_origen
y_pred = model.predict(X_origen)
model.predict(X_origen)
print("Miss Classification Loss: %f"%(1-accuracy_score(y_true, y_pred)))
```

Realizando una comparación con el dataset original, es posible observar que LDA tampoco se comporta muy bien. De hecho existen dos puntos azules mal clasificados, los que corresponden al 0.2% de error de clasificación.

### c) Clasificador QDA

Luego, se entrena otro clasificador para ver su comportamiento respecto a LDA. Este nuevo clasificador es Quadratic Discriminant Analysis (QDA), el cual a diferencia de LDA, no separa los conjuntos de datos o clases mediante una recta. Para lograr esto, QDA ya no toma el supuesto que todas las matrices de covarianzas de las clases son iguales. Así, al mantener la suposición de normalidad de los datos y asumir diferente matriz de covarianza para cada clase, se obtiene un término cuadrático en la frontera de clasificación.

<center> $$\begin{equation}
\delta_{k}(x) = - \frac{1}{2} log |\Sigma_k| - \frac{1}{2} (x - \mu_{k})^{T} \Sigma^{-1}_k (x - \mu_{k}) + log \pi_{k} \ \ \ \ \ (3) 
\end{equation}$$
</center>


```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
model = QDA()
model.fit(X,y)
visualize_border(model,X,y,"QDA")
```

```python
y_true = y
y_pred = model.predict(X)
model.predict(X)
print("Miss Classification Loss: %f"%(1-accuracy_score(y_true, y_pred)))
```

A través de la gráfica es posible visualizar que el comportamiento que tiene QDA no mejora mucho respecto a lo que es LDA. Se ve que la curva que separar los datos tiene una mayor tendencia hacia la derecha, intentando separa los datos ruidos que fueron inducidos. Sin embargo, QDA no pudo lograrlo separando casi la misma cantidad de datos que LDA. Lo anterior es posible verificar con el error de clasificación obtenido en este caso, que sería de 2.0%.

```python
model.fit(X_origen,y_origen)
visualize_border(model,X_origen,y_origen,"QDA sin ruido")
```

```python
y_true = y_origen
y_pred = model.predict(X)
model.predict(X_origen)
print("Miss Classification Loss: %f"%(1-accuracy_score(y_true, y_pred)))
```

Al igual que en el caso de LDA, se realiza una comparación del clasficador con el dataset original, sin ruido. De esto, es posible ver que QDA se comporta mejor que LDA, y se logra obtener resultados prácticamente perfectos, dado que separa ambos conjuntos de datos con un 0.0% de error de clasificación.

Ambas técnicas asumen normalidad de los datos, para beneficiarse de las propiedades matemáticas de esta distribución de probabilidad. Por esto mismo, ambos conjuntos de datos fueron normalizados antes de aplicar el algoritmo.

### e) Regresión Logística Regularizado con norma l2

En la regresión logística se busca maximizar la log-likelihood, en esta se asume una función sigmoidal para la probabilidad condicional, para esto se busca maximizar los parámetros $\beta$. La regularización l2 ($\lambda$) reduce los coeficientes $\beta$ proporcional a la norma L2 con una variable $\lambda$, imponiendo una penalización al valor de cada coeficiente. Hay buscar un valor de $\lambda$ eficiente, dado que si $\lambda$ es muy grande, el modelo puede sufrir underfitting o un bias elevado, mientras que un $\lambda$ demasiado bajo, puede no afectar al modelo en absoluto. 

La función a maximizar está dada por:

$$ \sum_{i=1}^{N} \left\{ y_i \beta^T x_i - log(1+e^{\beta^T x_i}) - \lambda \sum_{j=1}^{p} || \beta_{j}||_{2}^{2}  \right\}   $$

```python
from sklearn.linear_model import LogisticRegression as LR
def train_model(param):
    model = LR() #define your model
    model.set_params(C=param,penalty='l2')
    model.fit(X,y)
    y_true = y
    y_pred = model.predict(X)
    print("Miss Classification Loss: %f"%(1-accuracy_score(y_true, y_pred)))
    return model
p_min = 10**-3 #define your range
p_max = 1000#define your range
interactive(visualize_border_interactive,param=(p_min,p_max))
```

Con el widget interactivo de valores continuos es complejo trabajar por dos razones fundamentales. La primera tiene que ver con los valores que toma el parámetro de regularización, dado que al ir cambiando continuamente no se nota un cambio significativo en el resultado esperado en la gráfica. La segunda tiene que ver con el costo computacional, ya que si se entrega un rango de valores muy amplio para el parámetro se tendrían que entrenar muchos modelos, de los cuales no todos importan por lo mencionado en el punto anterior.
Es por esto que para las siguientes pruebas el parámetro será sintonizado mediante una lista, en la cual se mostrarán distintos valores de diferentes ordenes de magnitud, con los cuales se pueden ver mejor la incidencia del parámetro en el modelo.

```python
interactive(visualize_border_interactive,param=[10**i for i in range(-15,4)])
```

```python
import matplotlib.pylab as plt
from sklearn.metrics import precision_score

coefs = []
lambdas = [10**i for i in range(-15,7)]

for a in lambdas:
    model = LR()
    model.set_params(C=a,penalty='l2')
    model.fit(X,y)
    coefs.append(model.coef_)
    model.predict(X)
    ax = plt.gca()
#print coefs
names_regressors = ['Normal Multivariada','Senos y Cosenos']
for y_arr, label in zip(np.squeeze(coefs).T, names_regressors):
    plt.plot(lambdas, y_arr, label=label)
    
plt.legend()
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1]) # reverse axis
plt.title('Regresion Logistica Regularizada')
plt.xlabel('Parametro de regularizacion $\lambda$')
plt.ylabel('Coeficiente')
plt.legend(loc=2)
plt.show()
```

En este gráfico se representa la incidencia que tienen los coeficientes obtenidos para cada clase según el parámetro de regularización que se entrega al modelo. Es posible observar que a medida que $\lambda$ disminuye $\left ( 10^{-5} app.\right )$, los coeficientes ya tienen un valor prácticamente constante y cercano a 0. Es debido a esto que al verificar en el widget interactivo como se comporta la regresión logística en valores pequeños, no se notan cambios significativos en la frontera de decisión, teniendo un mismo error de clasificación entre $10^{-5}$ y $10^{-15}$.

### f) Soft Margin SVM Lineal


El modelo de SVM Soft-Margin está dado por:

$$ \mathop{Min}_{W,b,\epsilon_i > 0 }W^{T}W + C\sum_{i=1}^{n} \epsilon_i$$
$$ s.t $$
$$ y^{(i)} (W^{T}x^{(i)} + b) \leq 1 - \epsilon_i $$

El significado del parámetro C, determina el trade-off entre la maximización del margen y la minimización del error de entrenamiento. Una C pequeña hace que la frontera o hiperplano de decisión sea más suave, mientras que un parámetro más grande es más restrictivo haciendo que todos los ejemplos de entrenamiento se clasifiquen correctamente. 

Por el kernel utilizado en la SVM, la frontera a implementar es lineal por lo que se comporta de manera similiar a los algoritmos anteriormente analizados, LDA,QDA y regresión logística.

```python
from sklearn.svm import SVC as SVM #SVC is for classification
def train_model(param):
    model= SVM()
    model.set_params(C=param,kernel='linear')
    model.fit(X,y)
    y_true = y
    y_pred = model.predict(X)
    print("Miss Classification Loss: %f"%(1-accuracy_score(y_true, y_pred)))
    return model

p_min = 0.00001 #define your range
p_max = 100#define your range
interactive(visualize_border_interactive,param=(p_min,p_max))#use interactive
```

```python
interactive(visualize_border_interactive,param=[10**i for i in range(-4,6)])#use interactive
```

Es posible observar que para un kernel lineal en la SVM, el resultado obtenido es bastante bueno. Para los valores más bajos que se puede tomar, se ve que el error de clasificación es de un 1.04%, y a medida que se va aumentando el parámetro de regularización C, este error disminuye a 0.8% con un C=1. Si se evalúa con C>1 es posible visualizar, gracias al widget, que el clasificador no cambia.

(i) Construya un algoritmo k-NN para la clasificaci´on del problema. Mediante la imagen interactiva explore
diferentes valores del par´ametro k. Discuta el significado y efecto esperado de este par´ametro. Analice
cualitativamente lo observado.

El significado del parámetro entregado al modelo KNN, es la cantidad de vecinos utilizados en el algoritmo. En la clasificación k-NN, el resultado es una membresía de clase. Un objeto se clasifica por el voto mayoritario de sus vecinos, asignándose a la membresía de la clase más común entre sus k vecinos más cercanos (k es un número entero positivo, típicamente pequeño). Si k = 1, entonces el objeto simplemente se asigna a la clase de ese vecino más cercano.
